```python
import sys
sys.path.append('../semantic_spaces/')

import numpy as np
import pandas as pd
import embeddings as emb
```

```python
'''
Load GloVe
'''

G = emb.glove.load_embedding('../semantic_spaces/glove-german/w2v_50D.txt')
```

```python
'''
Retrieve starting point for our models
'''

X = np.zeros((len(G), len(G['kayak'])))
V = G.key_to_index.keys()
for i, v in enumerate(V): X[i,:] = G[v]
mu = X.mean(axis = 0)
del X, V

with open('./resources/model/priors.npy', 'wb') as f:
    np.save(f, mu)
```

```python
'''
Create a mini version including only the items
that we actually need to save some RAM & processing
time during experiment.
'''

M = pd.read_csv('./resources/stimuli.csv', encoding = 'utf-8')
V = ['mehl', 'mal', 'goethe', 'g√∂re', 'bund', 'mund', 'starkstrom', 'symptom', 'modern', 'ungern', 'walken', 'borgen', 'waten']

for i in np.arange(0, len(M), 1):
    p = M.loc[i].file.split('/')[-1].split('_')[0].lower().split('-')
    p0 = str(p[0].encode().decode())
    p1 = p[1]
    if p0 not in V: V.append(p0)
    if p1 not in V: V.append(p1)

X = np.zeros((len(V), len(G['kayak'])))
for i, v in enumerate(V): 
    X[i,:] = G[str(v)]

with open('./resources/model/vocab.npy', 'wb') as f:
    V = np.array([str(v) for v in V])
    np.save(f, V)

with open('./resources/model/vectors.npy', 'wb') as f:
    np.save(f, X)

emb.glove.numpy_to_word2vec(f_in_voc = './resources/model/vocab.npy', 
                            f_in_vec = './resources/model/vectors.npy',
                            f_out = './resources/model/w2v_50D.txt')
```
