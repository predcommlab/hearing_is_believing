```python
import numpy as np
import pandas as pd
import os, glob
```

```python
pids = []

for i, file in enumerate(glob.glob('./data/raw/*.csv')):
    N = pd.read_csv(file)
    
    indcs_t = np.where((N.trial_type == '2afc_audio_visual') & (N.no >= 0) & (N.is_control == False))[0]
    indcs_c = np.where((N.trial_type == '2afc_audio_visual') & (N.no >= 0) & (N.is_control == True))[0]
    pid = np.unique(N.loc[indcs_t].pid.tolist())[0]
    if pid in pids: print(f'----- detected duplicate pid {pid} -----')
    else: pids.append(pid)
    
    t = N.loc[indcs_t]
    c = N.loc[indcs_c]
    
    rts = np.array(t.rt.tolist())
    misses = np.isnan(rts).sum()
    rt_low = (rts < 125).sum()
    rt_min, rt_max = np.round(np.nanpercentile(rts, 2.5)), np.round(np.nanpercentile(rts, 97.5))
    
    options = np.array(t.choice_option.tolist())
    left = (options == 'left').sum()
    right = (options == 'right').sum()
    
    choices = np.array(t.choice_is_target.tolist())
    performance = np.round(choices.sum() / 240, 2)
    
    ctrl_ch = np.array(c.choice_is_target.tolist())
    ctrl_p = np.round(ctrl_ch.sum() / 24, 2)
    
    print(f'{i} [{pid}] rt={np.round(np.nanmean(rts))} {rt_min, rt_max}\tperf = {performance} {left, right}\tctrl = {ctrl_p}\tmisses = {misses+rt_low}')
```

```python
for i, file in enumerate(glob.glob('./data/raw/*.csv')):
    N = pd.read_csv(file)
    
    indcs_t = np.where((N.trial_type == '2afc_audio_visual') & (N.no >= 0) & (N.is_control == False))[0]
    indcs_c = np.where((N.trial_type == '2afc_audio_visual') & (N.no >= 0) & (N.is_control == True))[0]
    indcs_p = np.where((N.trial_type == 'word_categorisation_circle'))[0]
    indcs_s = np.where((N.trial_type == 'survey-multi-choice'))[0]
    pid = np.unique(N.loc[indcs_t].pid.tolist())[0]
    
    t = N.loc[indcs_t]
    c = N.loc[indcs_c]
    p = N.loc[indcs_p]
    
    choices = np.array(t.choice_is_target.tolist())
    performance = np.round(choices.sum() / 240, 2)
    recognition = np.round(np.array(p.correct.tolist()).astype(int).sum() / 60, 2)
    awareness = N.loc[indcs_s[0]].q_awareness
    no = N.loc[indcs_s[1]].q_number
    conflict = N.loc[indcs_s[2]].q_conflict
    strat = N.loc[indcs_s[3]].q_decision
    shift = N.loc[indcs_s[3]].q_strategy_shift
    reps = N.loc[indcs_s[4]].q_repetitions
    acoustics = N.loc[indcs_s[5]].q_repetitions_acoustic
    
    print(f'{i} [{pid}] perf = {performance}\treco = {recognition}')
    print(f'A = {awareness}\tN = {no}\tC = {conflict}\nS = {strat}\tQ = {shift}')
    print('-')
```

```python
# setup paths
f_T = '../../semantic_spaces/dumps/stimuli_pairs_from_old.xlsx'
f_C = '../../semantic_spaces/dumps/stimuli_pairs_control.xlsx'
f_P = '../../semantic_spaces/dumps/stimuli_pairs_practice.xlsx'

# setup path vars
path = './data/raw/*.csv'

import glob

if True:
    # load reference data
    N = pd.read_excel(f_T)
    
    # setup dummy data structures
    d_pid = []
    d_no = []
    d_time = []
    d_block = []
    d_kappa = []
    d_context1 = []
    d_context2 = []
    d_contexta = []
    d_contextb = []
    d_word_a = []
    d_word_b = []
    d_position_a = []
    d_face = []
    d_feature = []
    d_fit_a_LT = []
    d_fit_b_LT = []
    d_fit_a_LT_r = []
    d_fit_b_LT_r = []
    d_freq_a = []
    d_freq_b = []
    d_hist = []
    d_rt = []
    d_correct = []
    d_ismiss = []
    d_isunreasonable = []
    d_iscontrol = []
    d_strategy = []
    
    # loop over participants
    for i, f in enumerate(glob.glob(path)):
        pid = np.array(f.split('/'))[-1].split('_')[5]
        
        print(f'Merging dataset {pid}.')
        
        # load trial data
        M = pd.read_csv(f)
        T = M.loc[(M.trial_type == '2afc_audio_visual') & (M.no >= 0)]
        
        # setup vectors (where possible)
        t = np.array((T.no - T.no.mean()) / T.no.std())
        b = np.array(T.block.tolist())
        n = np.array(T.no.tolist())
        rt = np.array(T.rt.tolist())
        ismiss = np.isnan(rt)
        isunreasonable = rt < 125
        iscontrol = np.array(T.is_control.tolist()).astype(int)
        
        # setup dummies
        hist = []
        k = []
        LT_M = []
        LT_N = []
        c1 = []
        c2 = []
        cona = []
        conb = []
        c = []
        rep = []
        freq_a = []
        freq_b = []
        worda = []
        wordb = []
        positiona = []
        face = []
        feature = []        
        
        # loop over trials
        for (t1, t2, pos, no, stim, ch, img, ctrl) in zip(T.option_left, T.option_right, T.target_position, T.no, T.stimulus, T.choice_option, T.prime, iscontrol):
            # read repetition
            r = int(np.array(stim.split('_'))[-2][1:])
            
            # read kappa-based t/a (and correct misspelt items)
            kt, ka = stim.split('/')[-1].split('_')[0].split('-')
            if ka == 'wert': ka = 'Wert'
            if ka == 'warten': ka = 'waten'
            
            # code baseline t/a
            ct, ca = (t1, t2) if pos == 'left' else (t2, t1)
            hctca = f'{ct}_{ca}'
            rep.append(hctca in hist)
            hist.append(hctca)
            
            if not ctrl:
                # find index in reference
                indx = np.where((N.target1 == kt) & (N.target2 == ka))[0]

                # identify speaker context
                if ct == kt:
                    context1 = N.loc[indx].context1.tolist()[0] # speaker
                    context2 = N.loc[indx].context2.tolist()[0] # alternative
                else:
                    context1 = N.loc[indx].context2.tolist()[0] # speaker
                    context2 = N.loc[indx].context1.tolist()[0] # alternative

                # find semantic fits
                if ct == kt:
                    fa_LT = N.loc[indx].t1_to_c1.tolist()[0]
                    fb_LT = N.loc[indx].t2_to_c1.tolist()[0]
                else:
                    fa_LT = N.loc[indx].t1_to_c2.tolist()[0]
                    fb_LT = N.loc[indx].t2_to_c2.tolist()[0]
                    
                # find contexts
                con_a = N.loc[indx].context1.tolist()[0]
                con_b = N.loc[indx].context2.tolist()[0]

                freq_a.append(N.loc[indx].t1_F.tolist()[0])
                freq_b.append(N.loc[indx].t2_F.tolist()[0])

                # update priors
                no = int(no)+6

                # write kappa
                k.append(N.loc[indx][f'f{r}_k'].tolist()[0])
            else:
                context1, context2, con_a, con_b, fa_Lt, fb_LT = None, None, None, None, 0, 0
                freq_a.append(0)
                freq_b.append(0)
                k.append(0)
            
            # write fits
            LT_M.append(fa_LT)
            LT_N.append(fb_LT)
            
            # write contexts
            c1.append(context1)
            c2.append(context2)
            cona.append(con_a)
            conb.append(con_b)
            
            # write words
            worda.append(kt)
            wordb.append(ka)
            
            # write position a
            positiona.append('left' if t1 == kt else 'right')
            
            # write face & feature
            i_face, i_feature = img.split('/')[-1].split('.')[0].split('_')
            face.append(i_face)
            feature.append(i_feature)
            
            # write chose(word_a)
            if not ctrl: 
                if ch == 'left': c.append(t1 == kt)
                else: c.append(t2 == kt)
            else:
                # or write chose(correct) for control items
                c.append(ch == pos)
        
        # log raw fit values
        d_fit_a_LT_r = np.hstack((d_fit_a_LT_r, LT_M))
        d_fit_b_LT_r = np.hstack((d_fit_b_LT_r, LT_N))
        
        # z-score fit values
        LT_M = np.array(LT_M)
        LT_M = (LT_M - LT_M.mean()) / LT_M.std()
        
        LT_N = np.array(LT_N)
        LT_N = (LT_N - LT_N.mean()) / LT_N.std()
        
        # log z-scored fits
        d_fit_a_LT = np.hstack((d_fit_a_LT, LT_M))
        d_fit_b_LT = np.hstack((d_fit_b_LT, LT_N))
        
        # read out strategy
        X = pd.read_csv(f)
        s_indcs = np.where((X.trial_type == 'survey-multi-choice'))[0]
        q_strategy = X.loc[s_indcs].q_decision.tolist()[3]
        
        # log everything else
        d_pid = np.hstack((d_pid, [pid]*len(LT_M)))
        d_no = np.hstack((d_no, n))
        d_time = np.hstack((d_time, t))
        d_block = np.hstack((d_block, b))
        d_kappa = np.hstack((d_kappa, k))
        d_context1 = np.hstack((d_context1, c1))
        d_context2 = np.hstack((d_context2, c2))
        d_contexta = np.hstack((d_contexta, cona))
        d_contextb = np.hstack((d_contextb, conb))
        d_rt = np.hstack((d_rt, rt))
        d_correct = np.hstack((d_correct, c))
        d_strategy = np.hstack((d_strategy, [q_strategy]*len(LT_M)))
        d_hist = np.hstack((d_hist, rep))
        d_freq_a = np.hstack((d_freq_a, freq_a))
        d_freq_b = np.hstack((d_freq_b, freq_b))
        d_word_a = np.hstack((d_word_a, worda))
        d_word_b = np.hstack((d_word_b, wordb))
        d_position_a = np.hstack((d_position_a, positiona))
        d_face = np.hstack((d_face, face))
        d_feature = np.hstack((d_feature, feature))
        d_ismiss = np.hstack((d_ismiss, ismiss))
        d_isunreasonable = np.hstack((d_isunreasonable, isunreasonable))
        d_iscontrol = np.hstack((d_iscontrol, iscontrol))
    
    # create and save data frame
    df = pd.DataFrame.from_dict({'pid': d_pid, 'no': d_no, 'time': d_time, 'block': d_block,
                                 'kappa': d_kappa, 'fit_a': d_fit_a_LT, 'fit_b': d_fit_b_LT,
                                 'freq_a': d_freq_a, 'freq_b': d_freq_b,
                                 'rfit_a': d_fit_a_LT_r, 'rfit_b': d_fit_b_LT_r, 'repetition': d_hist,
                                 'context_a': d_contexta, 'context_b': d_contextb,
                                 'word_a': d_word_a, 'word_b': d_word_b, 'position_a': d_position_a,
                                 'speaker_context_tgt': d_context1, 'speaker_context_alt': d_context2,
                                 'speaker_face': d_face, 'speaker_feature': d_feature,
                                 'strategy': d_strategy,
                                 'ismiss': d_ismiss, 'isunreasonable': d_isunreasonable, 'iscontrol': d_iscontrol,
                                 'chose_a': d_correct, 'rt': d_rt})
    df.to_csv('./data/behaviour_raw.csv')
```
